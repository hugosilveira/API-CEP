```python
import requests
import pandas as pd

ceps = ["01001-000", "87005-000", "55291-010"]  # Adicione os CEPs desejados à lista

dados_enderecos = []

for cep in ceps:
    dados_get = requests.get(f"https://viacep.com.br/ws/{cep}/json/")
    dados_enderecos.append(dados_get.json())

df_enderecos = pd.DataFrame(dados_enderecos)

df_enderecos
```
